Foi feita uma reestruturao de algumas colunas da tabela e criao de outras. Alm disso, todas as tabelas foram unidas em apenas uma tabela geral.

Na anlise de evaso notou-se que cerca de 26% dos planos foram cancelados, no havendo grande relevancia entre gnero.





No entanto, notou-se uma grande quantidade de evaso dos clientes senior, onde mesmo sendo uma parcela muito menor no nmero total de evaso, pode ser um ponto de ateno.




Notou-se tambm que quanto maior o gasto mensal, maior  a porcentagem de evaso, sendo que a menor porcentagem  de 20% onde o gasto mximo  de R$ 45,00.

 

A quantidade de meses e o tipo de contrato tambm influeciam a taxa de permanencia, sendo que quanto mais tempo o cliente est assinando, menos chance ele tem de cancelar os servios.



Outra anlise til  a porcentagem de cancelamento por tipo de contrato, onde os contratos de maior durao tendem a ter menos cancelamentos e tambm a mdia de meses que as contas permanecem ativas aumenta conforme a durao de cada contrato

 

## Concluses

A anlise evidenciou uma taxa de evaso de 26%, sendo influenciada principalmente pelo perfil do cliente (idoso ou no), valor gasto mensalmente, tempo de permanncia na empresa e tipo de contrato.

Clientes idosos apresentam maior tendncia de cancelamento, mesmo representando uma menor fatia da base total. Clientes que gastam mais tambm tendem a cancelar mais, o que pode indicar problemas de percepo de valor. Alm disso, foi identificado que clientes com menor tempo de conta e contratos mensais so os que mais cancelam os servios, enquanto aqueles com contratos mais longos e maior tempo de relacionamento demonstram maior fidelizao.

Possveis solues para os problemas citados so:

* Oferecer incentivos para contratos mais longos, como descontos progressivos.
* Criar programas de fidelidade e benefcios para clientes seniores.
* Avaliar a percepo de custo-benefcio dos planos com gasto mensal elevado.
* Implementar aes de reteno nos primeiros 2 anos de conta, perodo mais crtico.
* Monitorar feedbacks de clientes que cancelam, especialmente entre os que tm contratos curtos.
