# Tratando os dados

```python
# verificando dados nulos

df.isnull().sum()
```

### tratando coluna Churn

```python
df.Churn.unique()
```

```python
Churn_null = df.query('Churn == ""').index
```

```python
df.drop(labels=Churn_null, axis=0, inplace=True)
```

```python
df.Churn.unique()
```

```python
df['Churn'] = df.Churn.apply(lambda x: 1 if x=='Yes' else 0)
df['Churn'] = df['Churn'].astype(bool)
df['Churn'].unique()
```

### Tratando coluna PaperlessBiling

```python
df['PaperlessBilling'] = df['PaperlessBilling'].apply(lambda x: 1 if x=='Yes' else 0)
df['PaperlessBilling'] = df['PaperlessBilling'].astype(bool)
df['PaperlessBilling'].unique()
```

### Tratando as colunas Charges.Total e Charges.Monthly

```python
valores_excluir_charges_total = df.query('`Charges.Total` == " "').index
```

```python
df.drop(valores_excluir_charges_total, axis=0, inplace=True)
```

```python
df['Charges.Total'] = df['Charges.Total'].astype(np.float64)
```

```python
df.rename(columns={'Charges.Total': 'ChargesTotal', 'Charges.Monthly': 'ChargesMonthly' }, inplace=True)
```
